```python
import pandas as pd

tabela  = pd.read_csv('archive/all_indices_data.csv')
display(tabela)
```

```python
tabela['resultado'] = tabela['close'] - tabela['open']
display(tabela)
```

```python
tabela = tabela[tabela['open'] != 0]
```

```python
tabela = tabela[tabela['volume'] > 10000]
display(tabela)
```

```python
tabela = tabela[tabela['resultado'] != 0.000000]
display(tabela)
```

```python
display(tabela.groupby("ticker").sum(numeric_only=True))
tabela_aux = tabela.groupby("ticker").sum(numeric_only=True)
```

```python
tabela_aux = tabela_aux.sort_values('resultado', ascending=False)
display(tabela_aux)
```

```python
display(tabela_aux['resultado'].count())
tabela_positiva = tabela_aux[tabela_aux['resultado'] >=0]
tabela_negativa = tabela_aux[tabela_aux['resultado'] <0]


display(tabela_positiva['resultado'].count())
display(tabela_negativa['resultado'].count())
```
